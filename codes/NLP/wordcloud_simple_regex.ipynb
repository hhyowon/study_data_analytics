```python
from wordcloud import WordCloud
```

```python
# 한글 문장 리스트
sentences = [
    "이 영화는 놀라운 것이었습니다! 흥미진진한 줄거리였어요.",
    "나는 영화 전체 동안 웃을 수 없었습니다. 정말 웃기다!",
    "이 영화의 시네마토그래피는 최고 수준입니다. 놀라운 시각 효과!",
    "나는 이 영화의 연기를 최상으로 생각하지 않았어요.",
    "감독은 이 책의 본질을 이번 영화에 완벽하게 담아냈습니다.",
    "진정한 걸작! 나는 확실히 다시 볼 것이다.",
    "특수 효과는 놀랍습니다. 시각적인 스펙터클!",
    "스토리는 좀 혼란스러웠지만, 연기는 뛰어났어요.",
    "나는 이 영화에 실망했습니다. 기대에 부응하지 못했어요.",
    "이 영화는 내가 몇 년 동안 본 최고의 영화 중 하나에요. 강력 추천!"
]
```

## 자연어 처리 위한 단어 생성

```python
text = ''.join(sentences)
text
```

```python
## 조사 빼기
# 는,이,은,를,한,가,...
import re
pattern = r'[는은이가을를만의과에]'
text_regex = re.sub(pattern=pattern, repl='', string=text) # params : 변경대상, 대상 대체, 어떤것을
text_regex
```

```python
from collections import Counter
text_regex_count = Counter(text_regex.split()) #list를 count 
text_regex_count
```

```python
# 한글 폰트 경로 설정 (Windows 환경에 맞게 경로 수정)
font_path = 'C:/Windows/Fonts/malgun.ttf'

wordCloud = WordCloud(font_path=font_path, background_color='white')
wordCloud.generate_from_frequencies(text_regex_count)
```

```python
wordCloud.words_
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordCloud)
plt.axis('off')
plt.show()
```
