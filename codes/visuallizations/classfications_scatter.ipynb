```python
class_1_x_value = [0,1,2,3,4]
class_1_y_value = [1,3,5,7,9]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=class_1_x_value,y=class_1_y_value)
```

```python
class_2_x_value = [1.5,2.5,3.5,4.5,5.5]
class_2_y_value = [3.5,5.5,7.5,9.5,11.5]
```

## 하나 plot 표시

```python
# 두 클래스 변수를 한 Plot에 표현 
plt.scatter(x=class_1_x_value,y=class_1_y_value)
plt.scatter(x=class_2_x_value ,y=class_2_y_value)
plt.show()
```

```python
# 두 클래스 변수를 한 Plot에 표현 
plt.scatter(x=class_1_x_value,y=class_1_y_value)
plt.scatter(x=class_2_x_value ,y=class_2_y_value)
plt.show()
```

### title, label, legend(제목, 레이블, 범례)

```python
plt.scatter(x=class_1_x_value,y=class_1_y_value,label='Class One')
plt.scatter(x=class_2_x_value ,y=class_2_y_value,label='Class Two')
plt.title('Classfication Plot')
plt.xlabel('x Axis') # plt.xlabel('x 축')
plt.ylabel('y Axis')  # plt.ylabel('y 축')
plt.show()
```
